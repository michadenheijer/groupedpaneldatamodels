```python
import bonhomme_manresa
```

```python
from linearmodels.datasets import wage_panel

wage_panel_df = wage_panel.load()

wage_panel_df.set_index(["nr", "year"], inplace=True)
y = wage_panel_df["lwage"]
x = wage_panel_df.drop(columns=["occupation", "lwage"])
display(x)
display(y)
```

```python
N = x.index.get_level_values(0).nunique()
T = x.index.get_level_values(1).nunique()
K = x.shape[1]
x = x.values.reshape(N, T, K).astype(float)
y = y.values.reshape(N, T, 1).astype(float)
x, y
```

```python
bonhomme_manresa.grouped_fixed_effects(y, x, 3, max_iter=10000, tol=1e-8, gfe_iterations=100)
```
